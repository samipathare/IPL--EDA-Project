##9. The toss winner, toss decision, winner in final matches

```python
toss_win = ipl.groupby(['toss_winner', 'toss_decision', 'winner']).size().reset_index(name='count')
a = toss_win.pivot_table(index='toss_winner', columns=['toss_decision'], values='count').fillna(0)
a.plot(kind='bar')
plt.xlabel('Toss Winner')
plt.ylabel('Count')
plt.title('Toss Winner, Decision, and Match Wins')
```

## 12. What is the probability of winning a match if the toss was won?

```python
winner = ipl[ipl['toss_winner'] == ipl['winner']]

total = ipl.shape[0]
toss_win_match = winner.shape[0]
probability= toss_win_match / total
plt.figure(figsize=(8, 6))
sns.barplot(x=['Probability of Winning Toss and Match', 'Probability of Losing Toss and Winning Match'],
            y=[probability, 1 - probability], palette='pastel')
plt.title('Probability of Winning a Match Based on Toss Result')
plt.ylabel('Probability')
#plt.show()
```

## 13.The total runs by four's hit and the total number of four's hit by each team


```python

fours=dlvry[dlvry['batsman_runs']==4]
f=fours.groupby('batting_team')['batsman_runs'].agg([('total runs by fours','count')]).reset_index()
f
f.plot(kind='bar',color='grey')
plt.xlabel('batting team')
plt.ylabel('number of fours')
plt.title('total runs by fours hit by each team')
```

## 14. The fours hit by players

```python
four= dlvry[dlvry['batsman_runs']==4]
l= four.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index()
l
m= l.iloc[:10].plot(x='batsman', y='four', kind='bar',color='cyan')
plt.xlabel('player name')
plt.ylabel('number of four')
plt.title('Four hit by player ')
```

## 15. The number of fours hit in each season

```python
fours_data = dlvry[dlvry['total_runs'] == 4]
fours_season = ipl.groupby('season').size()
fours_season.plot(kind='bar', color='red')
plt.xlabel('Season')
plt.ylabel('Number of Fours')
plt.title('Number of Fours Hit in Each Season')
plt.xticks(rotation=45)
plt.show()
```

##16.  The total runs by the sixes hit and the number of sixes hit by each team

```python
a =dlvry.groupby('batting_team')['batsman_runs'].agg([('total runs by six','count')]).reset_index()
a.plot(kind='bar',colormap='ocean')
plt.xlabel('Batting team')
plt.ylabel('Number of six')
plt.title('Total runs by the sixes hit and the number of sixes hit by each team')
```

## 17. The number of sixes hit in each season

```python
six = dlvry[dlvry['total_runs'] == 6]
six_season = ipl.groupby('season').size()
six_season.plot(kind='bar',color ='maroon')
plt.xlabel('Season')
plt.ylabel('Number of Sixes')
plt.title('Number of Sixes Hit in Each Season')
```

## 18. The top 10 leading run scorers in IPL

```python
top_score = dlvry.groupby('batsman')['batsman_runs'].sum()
top_score = top_score.sort_values(ascending=False).head(10)
top_score.plot(kind='bar',color = 'silver')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top 10 Leading Run Scorers in IPL')
```
